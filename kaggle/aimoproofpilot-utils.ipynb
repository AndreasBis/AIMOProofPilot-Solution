```python
import subprocess
import sys
from pathlib import Path

wheels_path = Path("/kaggle/working/wheels")
requirements_path = Path("/kaggle/working/kaggle_requirements.txt")

subprocess.run(
    [
        "mkdir",
        "-p",
        str(wheels_path),
    ],
    check=True,
)
```

```python
packages = [
    "vllm>=0.19.0",
    "numpy<=2.3.5",
    "trl",
    "transformers",
    "peft",
    "datasets",
    "pyarrow",
    "openai-harmony",
    "safetensors",
    "jupyter-client",
    "sympy",
    "networkx",
    "mpmath",
    "z3-solver",
    "pytest",
]

subprocess.run(
    [
        sys.executable,
        "-m",
        "pip",
        "download",
        "--dest",
        str(wheels_path),
        "--python-version",
        "3.12",
        "--only-binary=:all:",
        "--extra-index-url",
        "https://download.pytorch.org/whl/cu130",
        "--extra-index-url",
        "https://pypi.nvidia.com",
        *packages,
    ],
    check=True,
)
```

```python
installed_packages = subprocess.run(
    [
        sys.executable,
        "-m",
        "pip",
        "freeze",
    ],
    check=True,
    capture_output=True,
    text=True,
).stdout
requirements_path.write_text(installed_packages)
```
